```python
pd.DataFrame([ndata.iloc[11].values, ndata.iloc[20].values], columns=questions)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0).fit(ndata)
kmeans.labels_
```

```python
l0 = ndata[kmeans.labels_ == 0].mean()
l1 = ndata[kmeans.labels_ == 1].mean()
kdf = pd.DataFrame([l0.values, l1.values], columns=questions)
kdf
```

```python
del_columns = questions[kdf.values.max(axis=0) - kdf.values.min(axis=0) < 1.5].values
del_columns
```

```python
kdf.drop(columns=del_columns)
```

```python
kdf.drop(columns=del_columns).shape
```

```python
smalldata = ndata.rename(columns=questions).drop(columns=del_columns)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
clf = make_pipeline(StandardScaler(), LinearSVC())
clf.fit(smalldata, kmeans.labels_)
```

```python
clf.score(smalldata, kmeans.labels_)
```

```python
clf[1].coef_
```
